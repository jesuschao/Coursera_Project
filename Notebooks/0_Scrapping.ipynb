```python
def get_mysql(init):
    
    url1 = f"https://www.coursera.org/search?query=mysql&page={init}&index=prod_all_launched_products_term_optimization"
    driver.get(url1)

    curso =[]
    rating=[]
    students=[]
    nivel=[]
    universidad =[]

    for i in range(1,11):
        try:
            try:
                curso.append(driver.find_element_by_xpath(f"//*[@id='main']/div/div/div[1]/div/div/div/div/div/div[2]/div[2]/div/ul/li[{i}]/div/div/a/div/div[2]/h2").text)
            except:
                curso.append("None")
            try:
                rating.append(driver.find_element_by_xpath(f"//*[@id='main']/div/div/div[1]/div/div/div/div/div/div[2]/div[2]/div/ul/li[{i}]/div/div/a/div/div[2]/div[3]/div[1]/div[1]/div/span[1]").text)
            except:
                rating.append("None")
            try:
                students.append(driver.find_element_by_xpath(f"//*[@id='main']/div/div/div[1]/div/div/div/div/div/div[2]/div[2]/div/ul/li[{i}]/div/div/a/div/div[2]/div[3]/div[1]/div[2]/span/span").text)
            except:
                students.append("None")
            try:
                nivel.append(driver.find_element_by_xpath(f"//*[@id='main']/div/div/div[1]/div/div/div/div/div/div[2]/div[2]/div/ul/li[{i}]/div/div/a/div/div[2]/div[3]/div[2]/div/div[2]/span").text)
            except:
                nivel.append("None")
            try:
                universidad.append(driver.find_element_by_xpath(f"//*[@id='main']/div/div/div[1]/div/div/div/div/div/div[2]/div[2]/div/ul/li[{i}]/div/div/a/div/div[2]/div[1]/div/span").text)
            except:
                universidad.append("None")
        except:
            break

    mysql=pd.DataFrame()
    mysql['course'] =curso
    mysql['rating'] = rating
    mysql['students'] =students
    mysql['level'] = nivel
    mysql['university'] = universidad

    return mysql
```

```python
mysqlfinal =pd.DataFrame()
inicio = 1
while inicio < 7:    
    midf = get_mysql(inicio)
    midf['skills'] ="MySQL"
    mysqlfinal = pd.concat([mysqlfinal, midf])
    print(mysqlfinal.shape)
    inicio +=1
```
